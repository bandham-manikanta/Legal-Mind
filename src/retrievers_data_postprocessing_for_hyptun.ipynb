```python
import os
datapath = '/home/bandham/Documents/StonyBrook_CourseWork/Spring 2025/LLM-AMS692.02/Legal-Mind/src/evaluation_results/retrieved_documents_per_experiment'
filenames = os.listdir(datapath)
filepaths = []
for filename in filenames:
    if filename.endswith('_retrieved_docs_answers.jsonl'):
        continue
    filepaths.append(os.path.join(datapath, filename))
filepaths
```

```python
for filepath in filepaths:
    df_1 = pd.read_json(filepath)

    def combined_retrieved_docs(docs):
        text = ''
        for doc in docs:
            text += doc['text'] + " & \n\n"
        return text

    df_1['combined_retrieved_docs'] = df_1['retrieved_documents'].apply(combined_retrieved_docs)
    df_1.columns = [c.upper() for c in df_1.columns]

    dfc = pd.concat([df, df_1], axis=1)
    print(dfc[dfc['query_text'] != dfc['QUERY_TEXT']].shape)
    dfc = dfc[['query_id', 'query_text', 'relevant_doc_ids', 'answer', 'COMBINED_RETRIEVED_DOCS']]
    output_filepath = filepath.replace('.json', '_with_retrieved_docs_answers.jsonl')
    dfc.to_json(output_filepath, orient='records', lines=True)
    print(output_filepath)
```
